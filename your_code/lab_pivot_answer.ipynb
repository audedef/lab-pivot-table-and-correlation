#### Import all the libraries that are necessary.

#### Open the ` Employee.csv` file and name your dataset `employee`.

```python
employee = pd.read_csv("Employee.csv")
employee
```

#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1 : in k€
employee.groupby(['Department']).agg({'Salary':pd.Series.mean}).rename(columns = {'Salary':'mean salary in k €'})
```

```python
# Method 2
employee.pivot_table(index='Department',values='Salary', aggfunc = ['mean'])
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
employee.pivot_table(index= ['Department','Title'], values=['Salary'], aggfunc = ['mean'])
```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.
*Hint: Use Pandas doc(see References in README.md) to explore the pivot_table() arguments*

```python
employee.pivot_table(index= ['Title'], values=['Salary', 'Years'], aggfunc = ['mean'])
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
employee.pivot_table(index= ['Department','Title'], values=['Salary', 'Name'], aggfunc = ['mean', pd.Series.nunique])
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
employee.pivot_table(index= ['Department','Title'], values=['Salary'], aggfunc = ['median', 'min', 'max', 'std'])
```

```python
# as there are only 1 VP HR, 1 VP IT, and 1 IT associate, the St Deviation isn't computed for these people. It can't exist
```

#### Based on your comments, fill the missing values with an appropriate value.

```python
employee.pivot_table(index= ['Department','Title'], values=['Salary'], aggfunc = ['median', 'min', 'max', 'std'], fill_value=0)
# it should be working well with fill_value argument... (cf read.me document on pivot table)
```

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
# we consider that for 1 employee, the company spends the salary x 1,45 for the taxes and so on
# let's add a column "Spent on salary" in our dataframe

employee['Spent on salaries'] = employee['Salary']*1.45
employee.head()
```

```python
# then let's create the pivot table
employee.pivot_table(index= ['Department'], values=['Name','Spent on salaries'], aggfunc = [pd.Series.nunique, 'sum'])
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
employee.pivot_table(index= ['Department', 'Title'], values=['Years','Salary'], aggfunc = ['sum', 'mean'])
```
